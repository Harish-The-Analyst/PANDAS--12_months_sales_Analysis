```python
all_data = pd.read_csv("C:\\Users\\nitis\\Downloads\\all_months_data.csv")
all_data.head()
```

#### removing NaN values 

```python
#all_data.isna().sum()
all_data = all_data.dropna(axis=0)
all_data.head()
```

```python
all_data[['month','Day','year']] = all_data['Order Date'].str.split('/',n=3, expand = True)
all_data.head()

```

```python
all_data['Years'] = all_data['year'].str[0:2]
all_data.head()
```

```python
#all_data = all_data.drop("Order Date",axis=1)
columns = ['year']
all_data.drop(columns, inplace=True, axis=1)
#all_data.head()
#all_data.columns
#del all_data["('Day', 'month', 'year')"]
#all_data.drop(['(Day, month, year)'], inplace=True)
all_data.head()
```

```python
#temp_data = all_data[all_data['month'] == 'Order Date'] # finding "order date" in month col

all_data = all_data[all_data['month'] != 'Order Date'] #saving vals that dont have "order date" in month col

all_data['month'] = all_data['month'].astype('int32') # changing the dtype to int32
```

#### converting cols into numeric as they in string 

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

#### making sales col

```python
all_data['sales']=all_data['Quantity Ordered']*all_data['Price Each']
```

```python
all_data.head()
```

#### what City had the highest number of sales?

```python
#using apply() we can apply functions on specific columns.

all_data['city'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[2][:3])

all_data['cities'] = all_data['city']+' '+ '('+ all_data['City'] +')'

all_data = all_data.drop(columns="City")
all_data = all_data.drop(columns="city")

all_data.head()
```

```python
results = all_data.groupby('cities').sum()
results
```

#### what products are more often sold together?

```python
 #orders havin same "order id's are sold together"

df = all_data[all_data['Order ID'].duplicated(keep=False)] #shows only the duplictaed vals of 'order id' colm together

#made a col "grouped_prod" which will have joined the product names with ',' as per the duplicate "order ID"
df['grouped_prod'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

#keeping only the two cols in the df and dropping teh duplicates in them
df = df[['Order ID','grouped_prod']].drop_duplicates()

df.head(20)
```

```python
# from stackoverflow
#for counting and showing the combinations :)
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['grouped_prod']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2))) #passing "row_list" as iterator and we wnna get "2" prods bought together 
    
#count.most_common(10)

for key, value in count.most_common(10):
    print(key,value)
```
